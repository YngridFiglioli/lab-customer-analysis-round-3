```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

```python
#Describe DataFrame.

df1.describe() 
```

```python
#Show a plot of the total number of responses.

sns.displot(df1["Response"])
```

```python
#Show a plot of the response rate by the sales channel.

p=sns.countplot(data=df1, y="Sales Channel", hue="Response", palette=['r', 'g'])
```

```python
 #Show a plot of the response rate by the total claim amount.

v=sns.barplot(data=df1, y="Total Claim Amount", x="Response", palette=['r', 'g'])
```

```python
#Show a plot of the response rate by income.

f=sns.barplot(data=df1, y='Income', x='Response', palette=['r', 'g'])
```
